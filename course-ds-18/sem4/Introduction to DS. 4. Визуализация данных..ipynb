## Библиотека Matplotlib

### Виды графиков:
- plt.plot() - ломаная линия;
- plt.scatter() - маркер или точечное рисование;
- plt.bar()
- plt.hist()

- Подпись оси абсцисс OX -> plt.xlabel();
- Подпись оси ординат OY -> plt.ylabel();
- Легенда -> plt.legend()
- Деления на оси абсцисс OX -> plt.xticks()
- Деления на оси ординат OY -> plt.yticks()

### Scatter-plot
**Что это**: Точечная диаграмма <br />
**Для чего рисовать**: нужно показать распредление объектов по категориям, в пространстве и т.д. <br />

**Детали про функцию**
Основное: передать функции хотя бы значения оси ординат. <br />
После передачи данных для отрисовки можно задавать параметры для оформления графика

```python
plt.figure(figsize = (15,8))
plt.scatter(foodCons['RedMeat'], foodCons['Eggs'], s = 40, color = 'purple', marker = 'v')
# ('o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X')¶
plt.title('Scatter Plot of Red Meat and Eggs Consumption', fontsize = 15)
plt.xlabel('Red meat consumption', fontsize = 12)
plt.ylabel('Eggs consumption', fontsize = 12)
plt.show()
```

```python
plt.scatter(foodCons['RedMeat'], foodCons['Eggs'], s = 40, color = 'purple', marker = 'v')
plt.scatter(foodCons['RedMeat'], foodCons['Fish'], s = 40, color = 'yellow', marker = 'o')
plt.title('Scatter Plot', fontsize = 15, loc = 'left')
plt.xlabel('Red meat consumption', fontsize = 12)
plt.legend(loc = 0)
plt.show()
```

### Bar plot

**Что это**: Столбцовая диаграмма <br />
**Для чего рисовать**: нужно показать распределение объектов по категориям, проранжировать (по возрастанию, убыванию) <br />

**Детали про функцию**: <br />
При отрисовке нужно передавать в качестве параметра высоты столбцов и категории этих высот. В отличие от привычного экселя функция bar не умеет сама производить такие расчеты. 

```python
sumCons = foodCons.sum()
bars = foodCons.columns
plt.bar(bars, sumCons, width = 0.1)
plt.show()
```

На случай, если вам нужно нарисовать не просто один столбец на одну категорию, а сделать два столбца на категорию или стобец с наполнением, можно воспользоваться кодом ниже. 

Мы сами генерируем данные. Есть 5 групп (переменная N), в которых есть студенты женского и мужского пола. Хотим на одном графике отразить средние баллы мужчин и женщин в каждой из групп (кортежи menMeans и womenMeans). 

```python
N = 5

menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = list(range(N))    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

neww = []
for i in range(N):
    neww.append(ind[i] + width)

p1 = plt.bar(ind, menMeans, width, color = 'g')
p2 = plt.bar(ind, womenMeans, width, bottom = menMeans, color = 'r')
p3 = plt.bar(neww, womenMeans, width, color = 'grey')

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend((p1[0], p3[0]), ('Men', 'Women'))

plt.show()
```

### Histogram
**Что это**: Гистограмма, обманчиво похож на столбцовую диаграмму <br />
**Для чего рисовать**: нужно показать распределение объектов по частоте их появления <br />
Ось X разделена на бины - секции/категории. 
По оси Y отложена частота появления объекта в конкретном бине.  

**Детали про функцию**: <br />
Принимает как датасет полностью, так и отдельные его стобцы.

```python
plt.hist(foodCons.Eggs, bins = 20)
plt.title('Гистограмма')
plt.show()
```

### Pie Chart

**Что это**: Круговая диаграмма <br />
**Для чего рисовать**: нужно показать, какую часть от целого занимает секция <br />

**Детали про функцию**: <br />
Умеет работать с одним столбцом, т.е. надо выбирать, какие данные передаем на отрисовку. 

```python
labels = foodCons.index
plt.pie(foodCons.Eggs, labels = labels, autopct ='%1.1f%%', radius = 3)
plt.show()
```

### Boxplot
**Что это**: Ящик с усами <br />
**Для чего рисовать**: Это прекрасный график, который по всем столбцам показывает инфо об описательных статистиках <br />

**Детали про функцию**: <br />

В библиотеке pyplot она некрасива. 

```python
plt.boxplot(foodCons)
plt.show()
```

### Несколько графиков в поле
До сих пор выводятся по одному графику в поле. Если нужно несколько? <br />
Для таких целей используется функция subplots: <br />
fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (width, height))


```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (6, 6))

axes[0, 0].scatter(foodCons['RedMeat'], foodCons['Eggs'], s = 40, color = 'purple', marker = 'v')
axes[0, 0].set_title('Scatter: Red Meat and Eggs ')

axes[1, 1].scatter(foodCons['RedMeat'], foodCons['Fish'], s = 40, color = 'yellow', marker = 'o')
axes[1, 1].set_title('Scatter: Red Meat and Fish ')

axes[1, 0].boxplot(foodCons.Eggs)
plt.show()
```

Есть одно общее свойство у всех функций в библиотеке seaborn, если изучать документацию. <br />
Можно передавать три аргумента: x, y, data. X и y - это названия столбцов, а data - источник этих столбцов.

```python
import seaborn as sns
```

```python
sns.barplot(data = foodCons)
plt.show()
```

```python
sns.countplot(foodCons.Eggs, palette = 'PuBu')
plt.show()
```

```python
sns.distplot(foodCons.Eggs, kde = False)
```

## Расчёты
### Максимум
Определим максимальное количество одной позиции в заказе.

```python
q_max = retail.Quantity.max()
print(q_max)
```

### Минимум
Определим минимальное количество одной позиции в заказе. Добавим условие, что количество позиций в заказе должно быть больше 0, так как значения с минусом означают, что позиции из заказа были удалены / отменены / возвращены (исходя из описания датасета: "If this code starts with letter 'c', it indicates a cancellation."):

```python
q_min = retail.Quantity[retail.Quantity > 0].min()
print(q_min)
```

### Среднее
Определим среднее количество одной позиции в заказе:

```python
q_mean = retail.Quantity[retail.Quantity > 0].mean()
print(q_mean)
```

### Дисперсия
Посчитаем дисперсию:

```python
q_var = retail.Quantity[retail.Quantity > 0].var()
print(q_var)
```

### Стандартное отклонение
Посчитаем стандартное отклонение:

```python
q_std = retail.Quantity[retail.Quantity > 0].std()
print(q_std)
```

### Медиана
Найдём медиану:

```python
q_med = retail.Quantity[retail.Quantity > 0].median()
print(q_med)
```

### Нижняя квартиль
Найдём нижнюю квартиль:

```python
q_lq = retail.Quantity[retail.Quantity > 0].quantile(q = 0.25)
print(q_lq)
```

### Верхняя квартиль
Найдём верхнюю квартиль:

```python
q_uq = retail.Quantity[retail.Quantity > 0].quantile(q = 0.75)
print(q_uq)
```

### Коэффициент вариации
Найдём коэффициент вариации:

```python
q_coef = q_std / q_mean
print(q_coef)
```

## Прокомментируем полученные данные

```python
q_descr = [q_max, q_min, q_mean, q_med, q_var, q_std, q_coef, q_lq, q_uq]
ind = ['Максимум', 'Минимум', 'Среднее', 'Медиана','Дисперсия','Стандартное отклонение',
       'Коэффициент вариации', 'Нижняя квартиль', 'Верхняя квартиль']
q_descr = pd.DataFrame(q_descr, index = ind, columns = ['Позиции в чеке'])
q_descr
```
